```python
import requests
import zipfile
import io
import os
import pandas as pd
from datasets import load_dataset
```

```python
# URL to the zip file
zip_url = "https://huggingface.co/datasets/takala/financial_phrasebank/resolve/main/data/FinancialPhraseBank-v1.0.zip"

# Download the zip file
response = requests.get(zip_url)
response.raise_for_status() # Raise an exception for HTTP errors

# Extract the zip file
with zipfile.ZipFile(io.BytesIO(response.content)) as zf:
    zf.extractall("financial_phrasebank_data")

print("Dataset downloaded and extracted to 'financial_phrasebank_data/'")
print(os.listdir('financial_phrasebank_data/'))
```

```python
# Load the specific text file from the extracted data, specifying the encoding
dataset = load_dataset("text", data_files="financial_phrasebank_data/FinancialPhraseBank-v1.0/Sentences_AllAgree.txt", encoding='latin-1')

# Display basic information about the dataset
print(dataset)
```

```python
# Access a split of the dataset (e.g., 'train')
train_data = dataset['train']

# Display the first few examples
print(train_data[0])
print(train_data[1])

# You can convert it to a Pandas DataFrame for easier manipulation
df = pd.DataFrame(train_data)
display(df.head())
```
